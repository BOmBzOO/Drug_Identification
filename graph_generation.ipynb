## Check for DATA Integrity

```python
drop_law = []
for i in range(0, len(df_Valve)-1):
    if ((df_Valve.iloc[i]['VALVE'] != df_Valve.iloc[i+1]['VALVE']) & (df_Valve.iloc[i]['STEP'] == df_Valve.iloc[i+1]['STEP'])):
        drop_law.append(i)
drop_law
```

```python
df_Valve
```

```python
df_Valve = df_Valve.drop(index=drop_law)
df_Valve = df_Valve.reset_index()
df_Valve
```

```python
df_Valve['Start_cycle'] = df_Valve['UNIXSTAMP']
df_Valve['Fan_on'] = df_Valve['UNIXSTAMP'].shift(-1)
df_Valve['End_cycle'] = df_Valve['UNIXSTAMP'].shift(-2)
df_Valve = df_Valve[:-2]
df_Valve
```

```python
df_valve_edit = pd.DataFrame()
for i in range(0, len(df_Valve)-1):
    if ((df_Valve.iloc[i]['VALVE'] == df_Valve.iloc[i+1]['VALVE'])):
#     if (((i % 2) == 0) & (df_Valve.iloc[i]['VALVE'] != df_Valve.iloc[i+1]['VALVE'])):
#         print(i)
        df_valve_edit = df_valve_edit.append(df_Valve.loc[i])
df_valve_edit
```

```python
Condition = {
    0:'DPG',
    1:'Acetic',
    2:'Benzaldehyde',
    3:'Keras',
}
```

```python
sensors = ['TGS2600', 
           'TGS2602', 
           'TGS2603',
           'TGS2620',
           'TGS2610']
```

```python
duration = 9 #13min
```

```python
for i in tqdm(range(len(df_valve_edit))):
    for minute in range(duration):
        df_temp = df_TGS[(df_TGS['UNIXSTAMP'] > df_valve_edit.iloc[i]['Start_cycle']) & 
                         (df_TGS['UNIXSTAMP'] < df_valve_edit.iloc[i]['Start_cycle']+(minute+1)*60)]
        createDirectory('data/'+str(minute+1)+'min')
        
        path = current_path + '/' + 'data/' + str(minute+1)+'min/'
        file_name = Condition[df_valve_edit.iloc[i]['VALVE']]+'_'+str(minute+1)+'min_' + str(int(i/len(Condition))+1)
        
        plt.figure(figsize=(5,5))
        for idx, sensor in enumerate(sensors):
            df_temp_temp = df_temp[df_temp[sensor] != 0][[sensor, 'UNIXSTAMP']]
#             print()
#             print(df_temp_temp)
            try:
                df_temp_temp = df_temp_temp/df_temp_temp.iloc[0][sensor]
            except IndexError as e:
                print(df_temp_temp)
                print(e)
                print()
            plt.plot(df_temp_temp.index, df_temp_temp[sensor], linewidth=2)
        plt.ylim(0,3)
        plt.axis('off')
#         plt.title(file_name)
        plt.savefig(path+file_name, dpi=200)
```
